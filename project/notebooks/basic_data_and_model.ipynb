# Basic Model

```python
# This is stupid. Don't make models that take entire dataframes as input.
# df_board: dataframe containing all moves for a single position
# temperature: scaling parameter for softmax. 0 for uniform random; infinity for max
def predict_stockfish_softmax(df_board, temperature=1):
    # Take softmax of Stockfish evals
    stockfish_evals = df_board['move_stockfish_eval'].values
    values = stockfish_evals - np.max(stockfish_evals)
    p_unnormalized = np.exp(values * temperature)
    p = p_unnormalized / sum(p_unnormalized)
    return p
```

```python
predict_stockfish_softmax(df.groupby('board_id').get_group(1).reset_index(drop=True))
```

```python
# This is stupid too. Process the data properly instead of working with dataframes...
# df: dataframe containing all games/positions
# model: function accepting a dataframe as an input and returning a list of probabilities
def get_performance(df, model):
    num_correct = 0
    num_tried = 0
    log_likelihood = 0 
    
    # All of this is slow and bad
    df_boards = df.groupby('board_id')
    for (name, df_board) in df_boards:
        df_board_reset = df_board.reset_index(drop=True)
        
        p = model(df_board)
        prediction = np.argmax(p)
        
        correct = df_board_reset['move_played'][prediction]
        ll = np.log(p[df_board_reset['move_played']])
        
        num_tried += 1
        if correct:
            num_correct += 1
        log_likelihood += ll
    return (num_correct, num_tried, log_likelihood)
```

```python
# Try a bunch of different temperatures
from functools import partial
for temp in [0, 1e-5, 1e-4, 2e-4, 2.5e-4, 3e-4, 5e-4, 1e-3, 2e-3, 3e-3, 5e-3, 1e-2, 1e-1, 1]:
    model = partial(predict_stockfish_softmax, temperature = temp)
    print(temp, get_performance(df, model))
```
