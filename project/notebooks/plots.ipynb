```python
import csv
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
```

```python
matplotlib.rcParams['font.size'] = 8
matplotlib.rcParams['figure.dpi'] = 200
matplotlib.rcParams['figure.figsize'] = (4, 4)
```

# ELO per move

```python
with open('data/positions_raw.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    
    elo_list = []
    time_used_list = []
    
    for row in reader:
        elo_list.append(int(row['ELO_self']))
        if row['clock_used'] != '':
            time_used_list.append(int(row['clock_used']))
    
```

```python
# display(elo_dict)
print(min(elo_list))
print(max(elo_list))
print(elo_list[:10])
print(time_used_list[:10])
```

```python
bins = range(800, 2800+1, 200)
plt.figure(figsize=(5, 2))

plt.hist(elo_list, color='k', bins=bins, rwidth=0.8)
plt.xticks(bins)
plt.yticks(range(0, 20000+1, 5000))
plt.xlabel('ELO')
plt.ylabel('Moves')

plt.show()
```

```python
bins = [0, 1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]
plt.figure(figsize=(5, 2))

plt.hist(time_used_list, color='k', bins=bins, rwidth=0.8)
plt.semilogx()
plt.xticks(bins[1:], [str(x) for x in bins[1:]])
plt.yticks(range(0, 25000+1, 5000))
plt.xlabel('Seconds Used')
plt.ylabel('Moves')

plt.show()
```
